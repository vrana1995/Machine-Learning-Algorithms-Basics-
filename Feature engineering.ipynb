# Feature engineering on flight dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Feature engineering 
## Fetching data month year from date of journey

final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]

#alternate using lambda function

#final_df['Date']=final_df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
#final_df['Month']=final_df['Date_of_Journey'].apply(lambda x:x.split('/')[1])
#final_df['Year']= final_df['Date_of_Journey'].apply(lambda x:x.split('/')[2])
```

```python
final_df
```

```python
final_df['Date']= final_df['Date'].astype(int)
final_df['Month']= final_df['Month'].astype(int)
final_df['Year']= final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
del final_df['Date_of_Journey']
```

```python
final_df.head()
```

```python
## now for arrival time we only need the arrival time and not the date so

final_df['Arrival_Time']= final_df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
final_df['Arrival_Time']
```

```python
final_df.head()
```

```python
## now splitting arrival time into hour and minutes
final_df['Arrival_hour']= final_df['Arrival_Time'].apply(lambda x:x.split(':')[0])
final_df['Arrival_min']= final_df['Arrival_Time'].apply(lambda x:x.split(':')[1])
```

```python
final_df.head()
```

```python
## changing the type
final_df['Arrival_hour']= final_df['Arrival_hour'].astype(int)
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df
```

```python
## now splitting Departure time into hour and minutes
final_df['Dept_hour']= final_df['Dep_Time'].str.split(':').str[0]
final_df['Dept_min']= final_df['Dep_Time'].str.split(':').str[1]
final_df['Dept_hour']= final_df['Dept_hour'].astype(int)
final_df['Dept_min']=final_df['Dept_min'].astype(int)

del final_df['Dep_Time']
                                                 
```

```python
final_df.info()
```

```python
## Considering the stop made between will affect the price of the flight we can covert the stops into numerical data
final_df['Total_Stops'].unique()
```

```python
## we see that nan is there. Let see how many nan values
final_df['Total_Stops'].isnull().sum()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops']= final_df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4,'nan':2})
```

```python
final_df.head()
```

```python
del final_df['Route']
```

```python
final_df.info()
```

```python
## now for the duration we split it and try to convert it into one single type
final_df['Duration'].str.split(' ').str[0]
```

```python
final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df['Duration_Hour']= final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df.info()
```

```python
# changing datatype
final_df['Duration_Hour'].astype(int)
```

```python
#so we have error while converting into int data type we get error as for 6474 and 2660 are having duration as 5m so we drop them and then convert
final_df[final_df['Duration_Hour']== '5m']
```

```python
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df[final_df['Duration_Hour']== '5m']## we don't have nan valueds now
```

```python
final_df
```

```python
final_df['Duration_min']=final_df['Duration'].str.split('h ').str[1].str.split('m').str[0]
```

```python
final_df
```

```python
final_df.info()
```

```python
final_df['Duration_min'].astype(int)
```

```python
## here also we are gettimg nan values because of which we are not able to convert the data type.
final_df
```

```python
final_df['Duration_min']
```

```python
final_df[final_df['Duration_min'].isnull()]
```

```python
##now we are getting a nan value but we know nan is because there is no record of minutes for records.so wwe can replace nan with 0.
final_df['Duration_min']=final_df['Duration_min'].fillna(0)
```

```python
final_df
```

```python
final_df.info()
final_df['Duration_min']=final_df['Duration_min'].astype(int)
final_df['Duration_Hour']=final_df['Duration_Hour'].astype(int)
del final_df['Duration']
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
final_df['Airline']=labelencoder.fit_transform(final_df['Airline'])
final_df['Source']= labelencoder.fit_transform(final_df['Source'])
final_df['Destination']= labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']= labelencoder.fit_transform(final_df['Additional_Info'])
```
